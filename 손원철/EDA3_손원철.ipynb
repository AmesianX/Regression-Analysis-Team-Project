```python
feet_df = pd.DataFrame(feet_dict)
feet_df.tail()
```

```python
feet_df.shape
```

### 1-1. All Feet variables's describe + Data Type + Missing Value Ratio

```python
feet_df_desc = feet_df.describe().transpose()
feet_df_desc["data type"] = feet_df.dtypes
feet_df_desc["MV ratio"] = round(feet_df.isnull().sum()/feet_df.shape[0]*100,3)
feet_df_desc
```

### 1-2. All Feet variables's Not Missing Value Plot

```python
feet_MV_Columns = feet_df.columns[feet_df.isnull().any()].tolist()
msno.bar(feet_df[feet_MV_Columns],\
            figsize=(20,8),color="#34495e",fontsize=15,labels=True)
plt.show()
```

```python
merged_feet_df = feet_df
merged_feet_df["logerror"] = merged["logerror"]
merged_feet_df["parcelid"] = merged["parcelid"]
merged_feet_df = merged_feet_df.set_index("parcelid")
merged_feet_df.tail()
```

```python
fig, ax = plt.subplots(3,4, figsize = (20,15))

for idx, n in enumerate(feet_df.columns):
    if n == 'logerror':
        continue
    sns.regplot(x=n, y='logerror', data=merged, ax = ax[idx//4,idx%4])
    ax[idx//4, idx%4].set(yticklabels=[])
    ax[idx//4, idx%4].set(xticklabels=[])
    continue
```
